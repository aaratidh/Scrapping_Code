```python
link = "https://www.ubi.pt/en/courses"
driver.get(link)

first_part ="//span[@id=\'"
middle_part = "ConteudoHolder_LV_Graus_Texto_GrauLabel_"
last_part = "\']"
id_first_part ="//table[@id ='ConteudoHolder_LV_Graus_LV_Cursos_By_Grau_"
id_third_part ="_itemPlaceholderContainer_"
id_fifth_part ="']//tr/td/a"
course_link=[]
category=[]


for i in range(0,4):
    obtained_x_path = first_part+ middle_part+str(i)+last_part
    driver.find_element_by_xpath(obtained_x_path).click()
    id_x_path =id_first_part+str(i)+id_third_part+str(i)+id_fifth_part
    course =driver.find_elements_by_xpath(id_x_path)
    for co in course:
        driver.implicitly_wait(10) 
        ce = co.get_attribute('text')
        name_link = co.get_attribute('href')
        print(ce ,name_link )
        course_name.append(ce)
        course_web_site.append(name_link)
        if i == 0:
            category.append('1st Cycle Degrees')
        elif i == 1:
            category.append("Integrated Master's Degrees")
        elif i == 2:
            category.append('2nd Cycle Degrees')
        elif i == 3:
            category.append('3rd Cycle Degrees')
        
            
    
        
        


        
        
        
        
        
        
        
       
    
    
        
        



```

```python
for i in course_web_site:
    try:
        driver.get(i)
        driver.implicitly_wait(10)
        driver.find_element_by_xpath("//span[@id='ConteudoHolder_Repeater_Nome_Curso_curso_info1_0_LV_Info_Curso_0_ECTSLabel_0']").click()
        ect=driver.find_element_by_xpath("//span[@id='ConteudoHolder_Repeater_Nome_Curso_curso_info1_0_LV_Info_Curso_0_ECTSLabel_0']").text
        print("ECT mode: ",ect)
        duration.append('ect ,1 year= 60 ect')
        duration_time.append(ect)
    except NoSuchElementException:
        duration.append('unknown')
        duration.append(' ')
        pass
```

```python
for i in course_web_site:
    try:
        driver.get(i)
        driver.implicitly_wait(10)
        driver.find_element_by_xpath("//span[@id='ConteudoHolder_Repeater_Nome_Curso_curso_info1_0_LV_Info_Curso_0_UcRegimeLabel_0']").click()
        ect_s=driver.find_element_by_xpath("//span[@id='ConteudoHolder_Repeater_Nome_Curso_curso_info1_0_LV_Info_Curso_0_UcRegimeLabel_0']").text
        print("ECT mode: ",ect_s)
        study_mode.append(ect_s)
    except NoSuchElementException:
        print("error aayo pheri search gar")
        study_mode.append("")
        pass
```
